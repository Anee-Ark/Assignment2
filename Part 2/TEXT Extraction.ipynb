```python
from PyPDF2 import PdfReader

# Specify the path of the PDF file
pdf_path = "/Users/Aneesh/Downloads/Archive 2 2/2024-l1-topics-combined-2.pdf"

# Create a PdfReader object
reader = PdfReader(pdf_path)

# Create a list to store extracted text from each page
text_list = []

# Go through every page and get the text
for i in range(len(reader.pages)):
    page = reader.pages[i]
    text = page.extract_text()
    text_list.append(text)

    # Print the text if needed
    print(text)

# Specify the path for the output text file
output_txt_path = "PypdfTEXT(1).txt"

# Write the extracted text to a txt file
with open(output_txt_path, 'w', encoding='utf-8') as file:
    for text in text_list:
        file.write(text + '\n')

# Print a message indicating that the text has been saved
print(f"Extracted text has been saved to {output_txt_path}")
```

```python
from lxml import etree

def tei_to_text(tei_file_path, output_text_file):
    # Parse the TEI XML file
    tree = etree.parse(tei_file_path)
    root = tree.getroot()

    # Extract text content from TEI XML
    text_content = []
    for text_node in root.xpath("//text()"):
        text_content.append(text_node)

    # Save text content to a text file
    with open(output_text_file, 'w') as text_file:
        text_file.write(' '.join(text_content))

# Specify paths
tei_file_path = "2024-l1-topics-combined-2.pdf.tei (2).xml"
output_text_file = "TEXT(1).txt"

# Call the function
tei_to_text(tei_file_path, output_text_file)
```
